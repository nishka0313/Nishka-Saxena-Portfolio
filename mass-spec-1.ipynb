```python
import pandas as pd 
import matplotlib.pyplot as plt
```

<h3>Fraction 1</h3>

```python
df1 = pd.read_csv('/Users/nishkasaxena/Desktop/Mass Spec Data/Fluorinated porphyrin data/Order34671_FeTPFP2O_Al_F1_LR_0_L15_1.txt', sep=r'\s+', header=None, names=['m/z', 'intensity'])
df1["intensity_norm"] = df1["intensity"] / df1["intensity"].max() * 100
df1
```

```python
plt.figure(figsize=(13, 5))
plt.vlines(df1["m/z"], ymin=0, ymax=df1["intensity"], color="black", linewidth=0.8)
plt.xlabel("m/z")
plt.ylabel("Intensity")
plt.title("Mass Spectrum 1")
plt.xlim(750, 1400)
plt.tight_layout()
plt.show()
```

** there was a very small bump at ~2000, cut out because it was very minimal.
